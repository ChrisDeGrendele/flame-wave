```python
Nuclei = { 

    "h" : [1,3],                                                                                                                     
    "he" : [3, 4],                                                                                                                         
    "li" : [6, 7],                                                                                                                         
    #"be" : [7],   add this back later so it doesn't mess up dictionary...                                                                                                                         
    "be" : [9],                                                                                                                            
    "b" : [8, 11],                                                                                                                        
    "c" : [11, 14],                                                                                                                       
    "n" : [12, 15],                                                                                                                       
    "o" : [14, 19],                                                                                                                       
    "f" : [17, 21],                                                                                                                       
    "ne" : [17, 24],                                                                                                                       
    "na" : [19, 27],                                                                                                                       
    "mg" : [20, 29],                                                                                                                       
    "al" : [22, 31],                                                                                                                       
    "si" : [23, 34],                                                                                                                       
    "p" : [27, 38],                                                                                                                       
    "s" : [28, 42],                                                                                                                       
    "cl" : [31, 45],                                                                                                                       
    "ar" : [32, 46],                                                                                                                       
    "k" : [35, 49],                                                                                                                        
    "ca" : [36, 49],                                                                                                                      
    "sc" : [40, 51],                                                                                                                       
    "ti" : [41, 53],                                                                                                                       
    "v" : [43, 55],                                                                                                                       
    "cr" : [44, 58],                                                                                                                       
    "mn" : [46, 61],                                                                                                                       
    "fe" : [47, 66],                                                                                                                       
    "co" : [50, 67],                                                                                                                       
    "ni" : [51, 68],                                                                                                                       
    "cu" : [55, 69],                                                                                                                       
    "zn" : [57, 72],                                                                                                                       
    "ga" : [59, 75],                                                                                                                       
    "ge" : [62, 78],                                                                                                                       
    "as" : [65, 79],                                                                                                                       
    "se" : [67, 83],                                                                                                                       
    "br" : [68, 83],                                                                                                                       
    "kr" : [69, 87],                                                                                                                       
    "rb" : [73, 85],                                                                                                                       
    "sr" : [74, 84],                                                                                                                       
    "y" : [75, 87],                                                                                                                       
    "zr" : [78, 90],                                                                                                                       
    "nb" : [82, 90],                                                                                                                       
    "mo" : [83, 90],                                                                                                                       
    "tc" : [89, 91]                                                                                                                   
}
All_Nuclei = []
for Nuc, Num in Nuclei.items():
    if (len(Num)==1):
        All_Nuclei.append(Nuc + str(Num[0]))
    elif (len(Num) > 2):
        for i in Num:
            All_Nuclei.append(Nuc + str(i))
    else:
        for i in range(Num[0], Num[1]+1):
            All_Nuclei.append(Nuc + str(i))
All_Nuclei.append("n") #neutron
All_Nuclei.append("be7") #adding be7 back
assert len(All_Nuclei) == 495



library_file = '../results07070315'
mylibrary = pyna.rates.Library(library_file)
escn_tabular = ["f20--o20-toki","ne20--f20-toki","o20--f20-toki","f20--ne20-toki"]
mesa_495_lib = mylibrary.linking_nuclei(All_Nuclei, with_reverse=True)

from pynucastro import Nucleus
new_rates = []
#screens out duplicate rates that we include in the escn_tabular.
for rate in mesa_495_lib.get_rates():
    if rate.reactants == [Nucleus("f20")] and rate.products == [Nucleus("o20")]:
        continue
    elif rate.reactants == [Nucleus("ne20")] and rate.products == [Nucleus("f20")]:
        continue
    elif rate.reactants == [Nucleus("o20")] and rate.products == [Nucleus("f20")]:
        continue
    elif rate.reactants == [Nucleus("f20")] and rate.products == [Nucleus("ne20")]:
        continue
    else:
        new_rates.append(rate)
        
mesa_495 = pyna.networks.StarKillerNetwork(rates = new_rates, rate_files=escn_tabular)
mesa_495.plot()

mesa_495.write_network()
```
